```python
#Task 4: Token Analytics
import re

def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'http\S+|www\S+', '', text)
    text = re.sub(r'\S+@\S+', '', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    text = re.sub(r'[^a-z\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()

    words = text.split()
    cleaned_words = [w for w in words if len(w) > 2 or w in ["no", "not"]]

    return cleaned_words, " ".join(cleaned_words)


# Sentences
sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

# Analytics
for i, sentence in enumerate(sentences, 1):
    tokens, cleaned = preprocess_text(sentence)

    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))

    if total_tokens > 0:
        avg_length = sum(len(word) for word in tokens) / total_tokens
    else:
        avg_length = 0

    print(f"\nSentence {i}:")
    print("Original Text        :", sentence)
    print("Cleaned Tokens       :", tokens)
    print("Cleaned Sentence     :", cleaned)
    print("Total Tokens         :", total_tokens)
    print("Unique Tokens        :", unique_tokens)
    print("Average Token Length :", round(avg_length, 2))
    print("-" * 60)
```

```python
#Task 5: Frequency Analysis

import re
from collections import Counter

def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'http\S+|www\S+', '', text)
    text = re.sub(r'\S+@\S+', '', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    text = re.sub(r'[^a-z\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()

    words = text.split()
    cleaned_words = [w for w in words if len(w) > 2 or w in ["no", "not"]]

    return cleaned_words


# Sentences
sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

# Combine all tokens
all_tokens = []

for sentence in sentences:
    tokens = preprocess_text(sentence)
    all_tokens.extend(tokens)

# Frequency count
word_counts = Counter(all_tokens)

# Top 10 most frequent words
most_common = word_counts.most_common(10)

# Top 5 least frequent words
least_common = sorted(word_counts.items(), key=lambda x: x[1])[:5]

print("Top 10 Most Frequent Words:")
for word, count in most_common:
    print(word, ":", count)

print("\nTop 5 Least Frequent Words:")
for word, count in least_common:
    print(word, ":", count)
```

```python
#Task 7: Error Handling

import re

def preprocess_text(text):
    # Handle empty input or non-string
    if not text or not isinstance(text, str):
        return [], ""

    # Convert to lowercase
    text = text.lower()

    # Remove URLs
    text = re.sub(r'http\S+|www\S+', '', text)

    # Remove emails
    text = re.sub(r'\S+@\S+', '', text)

    # Remove numbers
    text = re.sub(r'\d+', '', text)

    # Handle repeated characters
    text = re.sub(r'(.)\1{2,}', r'\1', text)

    # Remove special characters (including emojis)
    text = re.sub(r'[^a-z\s]', '', text)

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # If text becomes empty after cleaning
    if text == "":
        return [], ""

    # Remove short words (≤2), keep "no", "not"
    words = text.split()
    cleaned_words = [w for w in words if len(w) > 2 or w in ["no", "not"]]

    # Final check
    if not cleaned_words:
        return [], ""

    return cleaned_words, " ".join(cleaned_words)


def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, clean_text = preprocess_text(text)

        all_tokens.extend(tokens)
        clean_sentences.append(clean_text)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }


#  Test Cases for Error Handling
test_cases = [
    "",                 # Empty string
    "😍😍😂😂🔥🔥",     # Only emojis
    "1234567890",       # Only numbers
    "!!!@@@",           # Only symbols
    "   ",              # Only spaces
    None                # None input
]

for i, test in enumerate(test_cases, 1):
    tokens, cleaned = preprocess_text(test)
    print(f"\nTest Case {i}:")
    print("Input            :", test)
    print("Tokens           :", tokens)
    print("Cleaned Sentence :", cleaned)
    print("-" * 50)
```

1. What is the difference between "Love" and "love" in NLP?
In NLP, "Love" and "love" are treated as two different words if we don’t do preprocessing like lowercasing. "Love" (capital L) may come at the beginning of a sentence or be used as a proper noun, while "love" (small l) is a normal word. But in most cases, both have the same meaning. So we convert everything to lowercase to avoid treating them as different words unnecessarily.

2. What happens if stopwords are not removed?
If stopwords are not removed, the dataset will contain many common words like "is", "the", "and", etc. These words do not add much meaning but increase the size of the data. This can make the model slower and sometimes less accurate because it may focus on unimportant words instead of meaningful ones.

3. Provide two real-world scenarios where removing stopwords can be harmful.

In sentiment analysis, removing stopwords like "not" can change the meaning. For example, "not good" becomes "good", which gives the wrong sentiment.
In question answering systems, stopwords like "who", "what", "where" are important to understand the question. Removing them can make the question unclear.

4. Explain the difference between stemming and lemmatization.
Stemming is a process where words are reduced to their root form by simply cutting off endings. For example, "running" becomes "run" or sometimes "runn". It is faster but not always accurate.
Lemmatization is a more advanced process that converts a word to its correct base form using vocabulary and grammar rules. For example, "running" becomes "run" correctly. It is more accurate but takes more time.
